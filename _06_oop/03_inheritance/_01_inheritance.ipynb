# 상속 Inheritance
- 부모클래스 속성/메서드를 자식클래스에서 선언없이 물려받아 사용하는 것
- 중복된 속성/메서드를 재작성할 필요가 없다.
- 자식클래스를 부모클래스로 묶어 관리할 수 있다.
- 파이썬 다중상속 지원

```python
class Person:
    def __init__(self, name: str, phone: str) -> None:
        self.__name = name
        self.__phone = phone

    def hello(self):
        print(f'안녕하세요, 저는 {self.__name}입니다. 제 연락처는 {self.__phone}이에요.')

    def get_name(self):
        return self.__name

    def get_phone(self):
        return self.__phone

    # __str__(): print(객체)를 했을 때 어떤 문자열을 보여줄지 정하는 메서드
    def __str__(self):
        return f'Person(name={self.__name}, phone={self.__phone})'


p1 = Person('홍길동', '01012341234')
print(p1)
```

```python
# Person 상속 받은 자식 클래스 Employee 생성
class Employee(Person):
    def __init__(self, name: str, phone: str, company: str) -> None:
        super().__init__(name, phone)
        self.__company = company

    # 보모로 부터 상속받은 hello메소드 재정의
    def hello(self):
        super().hello()
        print(f'{self.__company}에서 일하고 있습니다')

    def __str__(self):
        super_str = super().__str__()
        return f'Person(name={self.get_name()}, phone={self.get_phone()}, company={self.__company})'

    def get_company(self):
        return self.__company

    def set_company(self, company: str):
        self.__company = company


em1 = Employee('홍길동', '01011111111', 'SK')
print(em1.get_name())
print(em1.get_phone())
em1.hello()
print(em1)

```

```python
class Developer(Person):
    def __init__(self, name: str, phone: str, *langs: str) -> None:
        super().__init__(name, phone)
        self.__langs = list(langs)

    # langs 언어 추가
    # 있다면 "이미 추가 되어있습니다", 없으면 추가
    def add_langs(self, *langs: str):
        for lang in langs:
            if lang in self.__langs:
                print(f"{lang} : 존재합니다")
                continue
            self.__langs.append(lang)
            print(f'{lang} 추가')

    # langs 언어 제거
    # 있으면 제거, 없으면 "존재하지 않는 언어입니다"
    def remove_langs(self, *langs: str):
        for lang in langs:
            if lang not in self.__langs:
                print(f"{lang} : 존재하지 않는 언어입니다")
                continue
            self.__langs.remove(lang)
            print(f'{lang} 삭제')

    #hello 오버라이딩
    # 부모의 hello 출력
    # 사용가능한 언어 목록 출력
    def hello(self):
        super().hello()
        print(f'{self.__langs} 언어를 사용할 수 있습니다')

    # __srt__ 오버라이드
    def __str__(self):
        super_str = super().__str__()
        return f'Developer(name={self.get_name()}, phone={self.get_phone()}, langs={self.__langs})'


dev = Developer("김뀨뀨", "010-7777-7777", "c")
print(dev)
dev.hello()

# 'c', 'java', 'c#', 'python'
dev.add_langs('c#', 'python', 'go', "Lua", "c#", "c#", 'java')
print(dev)
# dev.lang_check('c#')

dev.remove_langs('Lua','go','c++')
print(dev)
dev.hello()

```

## isinstance(), issubclass(), MRO

상속 관계에서는 객체가 어떤 클래스의 인스턴스인지, 클래스가 어떤 부모를 상속하는지 확인할 수 있다.

- `isinstance(객체, 클래스)`: 객체가 해당 클래스 또는 하위 클래스의 인스턴스인지 확인
- `issubclass(자식클래스, 부모클래스)`: 클래스 사이의 상속 관계 확인
- MRO(Method Resolution Order): 메서드를 찾는 순서


```python
print(f'클래스 확인 : {isinstance(dev, Developer)}' )
print(f'클래스 확인 : {isinstance(dev, Person)}' )
print(f'클래스 확인 : {isinstance(dev, Employee)}' )

print('-'*30)
# 첫번재 인수가 두번째 인수의 자식이냐?
# print("issubclass(Employee, Person) : " + issubclass(Employee, Person))

# object 클래스 == 모든 클래스의 부모
# class Test == class Test(Object)

```

## Python의 오버라이딩과 오버로딩 주의

Python은 자식 클래스에서 부모 클래스의 메서드를 다시 정의하는 오버라이딩은 자연스럽게 지원한다.
다만 Java처럼 같은 이름의 메서드를 매개변수 조합별로 여러 개 선언하는 메서드 오버로딩은 지원하지 않는다.
같은 클래스 안에서 같은 이름의 메서드를 다시 선언하면, 뒤에 선언한 메서드가 앞의 메서드를 덮어쓴다.
