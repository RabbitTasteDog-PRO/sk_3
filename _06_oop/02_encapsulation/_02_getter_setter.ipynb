## Getter/Setter 메서드
캡슐화된 클래스에서 속성을 제어하기 위한 메서드로써 기본적으로 다음 두 유형이 있다.
1. getter 속성의 값을 읽어내기 위한 메서드
2. setter 속성의 값을 설정하기 위한 메서드

```python
class MyClass:
    def __init__(self, value):
        self.__value = value

    # getter
    # 외부에서 private 속성 값을 얻어 가게하는 메소드
    def get_value(self):
        return self.__value

    # setter
    # 외부에서 private 값을 지정하는 메소드
    def set_value(self, value):
        self.__value = value


my_cls = MyClass(100)
print('my_cls.get_value() get 확인', my_cls.get_value())
my_cls.set_value(10000)
print('my_cls.get_value() set 확인', my_cls.get_value())

```

## Python의 `@property`

Python에서는 Java식 `get_value()`, `set_value()` 메서드도 사용할 수 있지만, 더 Python다운 방식으로는 `@property`를 자주 사용한다.

`@property`를 사용하면 메서드로 값을 검증하면서도, 사용하는 쪽에서는 속성처럼 읽고 쓸 수 있다.


```python
class Score:
    def __init__(self, score: int):
        self.__score = score

    # getter, setter를 오버로딩으로 사용할때 사용되는 프로퍼티
    @property
    def value(self):
        return self.__score

    @value.setter
    def value(self, score: int):
        if not (0 <= score < 100):
            raise ValueError('점수는 0 ~ 100 사이만 set 가능합니다')
        self.__score = score


score = Score(80)
print(f'socre : {score.value}')

# 프로퍼티설정되면 변수처럼 사용
score.value = 100
print(f'socre : {score.value}')

# score.value = 120
```
