# 캡슐화
- 속성/메서드를 하나의 클래스에 함께 선언
- private 속성(변수) 선언해서 외부에서 직접 접근을 막고, 제공된 메소드를 통해서 사용하도록 강제
  - private 변수 선언 방법 : 변수명 앞에 __를 붙인다.
- 캡슐화 원칙을 잘 지켰다면, 주요한 데이터가 외부에 노출되지 않은 은닉화 효과까지 얻게된다.

```python
class Account:
    def __init__(self, username, balance):
        # __가 붙은 속성 == 생성된 인스턴스만 사용 가능한 private 속성
        self.__username = username
        self.__balance = balance

    # 외부에서 private 속성을 다룰 수 있는 메소드
    def info(self):
        print(f'{self.__username}님의 현재 잔액은 {self.__balance:,}원 입니다')

    # 입금
    # 0 or 음수는 존재할수없다
    def deposit(self, money):
        if money <= 0:
            print(f'입금액이 유요하지 않습니다')
        else:
            self.__balance += money
            print(f'입금액: {money}원, 잔액 : {self.__balance:,}원 입니다')

    # 출금
    def withdraw(self, money):
        if (self.__balance - money) <= 0:
            print(f'잔액이 부족합니다.')
        elif money <= 0:
            print(f'출금액이 유요하지 않습니다')
        else:
            self.__balance -= money
            print(f'출금액: {money}원, 잔액 : {self.__balance:,}원 입니다')

    def get_user_name(self):
        return self.__username

    def get_user_balance(self):
        return self.__balance

```

```python
acc1 = Account("최지흠", 100_000_000_000_000)
print(acc1.get_user_name())
print(acc1.get_user_balance())

acc1.deposit(10000)
acc1.withdraw(10000)
acc1.deposit(0)
acc1.withdraw(0)

acc1.info()

# acc.withdraw(10_000)

# acc.deposit(50000)


```

## Python private과 name mangling

Python에서 `__balance`처럼 언더스코어 두 개로 시작하는 속성은 외부에서 그대로 접근하기 어렵게 이름이 바뀐다.
이를 name mangling이라고 한다.

다만 이것은 보안을 위한 완전한 접근 제한이라기보다, 클래스 외부에서 직접 사용하지 말라는 강한 신호에 가깝다.
