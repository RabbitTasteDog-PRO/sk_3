# for 반복문

- `for` 반복문은 iterable 객체의 요소를 하나씩 꺼내 실행한다.
- `str`, `list`, `tuple`, `dict`, `set`, `range` 등이 대표적인 iterable 객체이다.
- 반복 횟수가 정해져 있거나 컬렉션의 모든 요소를 순회할 때 자주 사용한다.

## 1. 문자열, 리스트, 튜플 순회

문자열은 문자 단위로, 리스트와 튜플은 요소 단위로 순회한다.

```python
text: str = 'hello world'
for ch in text:
    print(ch)

lst: list[int] = [10, 20, 30]
for num in lst:
    print(num)

tpl: tuple[int] = (30, 35, 40)
for num in tpl:
    print(num)
```

## 2. dict와 set 순회

`dict`를 그대로 순회하면 key가 나온다. `items()`를 사용하면 key와 value를 함께 꺼낼 수 있다.

`set`은 순서가 없는 자료형이므로 출력 순서는 실행 환경에 따라 달라질 수 있다.

```python
from typing import Any

# 어떤 타입이던 작성 가능 Any == Object
user_info: dict[str, Any] = \
    {
        'name': '홍길동',
        'age': 20,
        'hobby': ['독서', '유튜브']
    }

print('----- dict key 순회')
for key in user_info:
    print(f'key 만 따오기 : {key}, value : {user_info[key]}')

for dic in user_info.keys():
    print(f'키 순회 : {dic}')

for key, value in user_info.items():
    print(f'item unpacking // {key} : {value}')

for item in user_info.items():
    print(f'아이템 순회 : {item}')

print('---- set 순회 : if 추가해서 -----')
nums: set[int] = {10, 2, 3, 4, 3, 3, 5, 3430, 4, 5, 340324, 503}

for num in nums:
    print(f'nums : {num}')

results: list[int] = []

# 짝수 체크
for num in nums:
    if num % 2 == 0:
        results.append(num)
        # print(f'쭉수 체크 num : {num}')
print(results)
```

## 4. 중첩 반복문

반복문 안에 반복문을 넣으면 행과 열처럼 2차원 구조를 만들 수 있다. 구구단과 별찍기 같은 예제에 자주 사용한다.

```python
print('---- 중접 반복문 ----')
# 2중포문
for i in range(2, 10):
    for j in range(1, 10):
        print(f'{i} * {j} = {(i) * (j)}')
    print('-' * 20)  # 문자열은 곱하기로 반복출력
```

```python
print('---- 2단만 ----')
const: int = 2
for i in range(1, 10, 1):
    print(f'{const} * {i} = {const * i}')
```

```python
print('---- 구구단 : zero bsse ----')
# 2중포문
for i in 9:
    for j in range(0, 9):
        print(f'{i + 2} * {j + 1} = {(i + 2) * (j + 1)}')
```

```python
print('---- range를 이용한 반복횟수 설정 ----')
print("닭다리 3행 5열")
for row in range(3):
    for cul in range(5):
        print("🍗", end='')  #end 속성 : 프린트 종료 후 출력할 문자 지정
    print()  # 줄바꿈
```

## 5. zip과 enumerate

- `zip()`은 여러 iterable을 같은 순서끼리 묶는다.
- `enumerate()`는 인덱스와 요소를 함께 꺼낸다.

```python
print('---- zip 객체 ----')
names: list[str] = ['a', 'b', 'c']
nums: list[int] = [1, 2, 3]

z = zip(names, nums)
print(f'zip : {z}')

for name, num in z:
    print(f'name : {name}, num : {num}')

print('---- enumerate 객체 ----')
foods: list[str] = ['제육', '피자', '치킨', '훠거']

for index, food in enumerate(foods):
    print(f'index : {index}, food : {food}')
```

## 6. 반복문으로 평균 구하기

반복문을 사용하면 리스트나 딕셔너리 리스트에 들어 있는 값을 누적해서 평균을 계산할 수 있다.

```python
print('---- list에 담긴 평균 구하기 ---')
scores: list[int] = [100, 20, 50, 60, 99]

score_sum = 0
for score in scores:
    score_sum += score

score_avg = score_sum / len(scores)
print(f'평군 : {score_avg}')
# 숫자로만 이루어진 자료구조일 경우에는 sum함수 가능
# 딕셔너린 불가
print(f'평군 : {sum(scores) / len(scores)}')
```

```python
print('---- dictionary에 담긴 평균 구하기 ----')
students: list[dict[str, object]] = [
    {'name': '홍길동', 'score': 100},
    {'name': '신사임당', 'score': 77},
    {'name': '이순신', 'score': 85},
]

stu_sum: int = 0;
for student in students:
    # print(f'name : {student["name"]}, score : {student["score"]}')
    stu_sum += student.get('score', 0)
print(f'딕셔너리 평균 :  { round(stu_sum / len(students))}')

rng_sum: int = 0;
for index in range(len(students)):
    rng_sum += students[index].get('score', 0)
print(f'레인지를 활용한 평균 : {round(rng_sum / len(students))}')
```
