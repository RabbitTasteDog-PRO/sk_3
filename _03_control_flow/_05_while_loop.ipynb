# while 반복문

- `while` 반복문은 조건식이 `True`로 평가되는 동안 코드를 반복 실행한다.
- 조건식이 `False`가 되면 반복을 중지한다.
- 반복 횟수가 미리 정해져 있지 않고, 특정 조건이 만족될 때까지 반복해야 할 때 자주 사용한다.

## 1. 기본 while 구조

`while`은 보통 초기값, 조건식, 증감식을 함께 사용한다.

```python
i = 1
while i <= 3:
    print(i)
    i += 1
```

````python
from zmq import select

# 1~ 10 출력
print('---- 1~ 10 출력 ---- ')
while i <= 10:
    print(i)
    i += 1

# 1~ 10 역순
print('---- 1~ 10 역순 ---- ')
i = 10
while i >= 1:
    print(i)
    i -= 1
````

## 2. break로 반복 중지하기

`while True`는 조건식이 항상 참인 반복문이다. 이 구조를 사용할 때는 반드시 반복을 빠져나올 수 있는 `break` 조건을 만들어야 한다.
- break는 가까운 반복문을 종료한다

````python
print('--- i ~ 10 출력 ----')
i = 0;
while True:
    print(i)
    i += 1
    if i > 10:
        print('------ break -----')
        break

````

### 03-02-01. break

- 조건문 또는 반복문의 실행 구문에서 코드가 break를 만나면 해당 블록의 실행이 종료된다.

    (break 이후의 실행 구문이 실행되지 않고 종료되어 블록 바깥의 하단 코드가 실행된다.)


````python
for x in [1, 2, 3]:

    print("앞", x)

    if x == 2:
        print("if", x)
        print("끝")
        break

    print("뒤", x)
````

### 03-02-02. continue

- 반복문의 실행 구문에서 코드가 continue를 만나면 continue 이후의 블록 내 코드는 더 실행되지 않고 바로 다음 반복으로 넘어간다.

    (일반적으로 continue는 반복문의 특정 순간을 생략하고 진행하기 위해 사용한다.)

````python
for x in range(3):
    print("앞", x)

    if x == 2:
        print("if", x)
        print("continue 등장!")
        continue

    print("뒤", x)
````

## 3. 메뉴 주문 예제

````python
menu = """
=========
구내식당 메뉴
=========
1. 제육볶음
2. 김밥
3. 김치찌개
4. 된장찌개
0. 종료
"""

print(menu)

orders: list[str] = []

while True:
    try:
        sel = int(input("메뉴를 선택하세요"))
    except ValueError:
        raise ValueError("숫자만 입력하세요")

    print(f'선택된 번호 : {sel}')
    match sel:
        case 1:
            orders.append("제육볶음")
        case 2:
            orders.append("김밥")
        case 3:
            orders.append("된장찌개")
        case 4:
            orders.append("김치찌개")
        case 0:
            orders.append("종료")
            print('메뉴선택 종료')
            break
        case _:
            print("메뉴에 작성된 번호만 입력하세요")
    print(f"저장된 리스트 : {orders}")




````
