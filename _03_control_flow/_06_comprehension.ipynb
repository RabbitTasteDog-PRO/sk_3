# Comprehension 내포

- Comprehension은 한 줄짜리 반복문으로 새로운 컬렉션을 만드는 문법이다.
- 대표적으로 list comprehension과 dict comprehension이 있다.
- 반복문으로 값을 하나씩 추가하는 패턴을 더 짧고 읽기 좋게 표현할 수 있다.

- list comprehension
```
# list 생성
[표현식 for 변수 in 반복가능한객체]

# 조건 포함
[표현식 for 변수 in 반복가능한객체 if 조건식]
```

- dict comprehension
```
# dict 생성
{key표현식: value표현식 for 변수 in 반복가능한객체}

# 조건 포함
{key표현식: value표현식 for 변수 in 반복가능한객체 if 조건식}
```



## 1. for 반복문과 list comprehension 비교

먼저 일반 `for` 반복문으로 리스트를 만든 뒤, 같은 결과를 list comprehension으로 바꿔 본다.

````python
from random import random

print('----- 일반 리스트 반환 ------')


def make_numbers_whit_for() -> list[int]:
    numbers: list[int] = []
    for n in range(1, 6):
        numbers.append(n)
    return numbers


print(f'리스트 반환 : {make_numbers_whit_for()}')

print('----- 컴프리핸선으로 리스트 반환 ------')


def make_numbers_whit_comprehension() -> list[int]:
    return [lst * 2 for lst in range(1, 6)]


print(f'컴프리핸션 반환 : {make_numbers_whit_comprehension()}')
````

## 2. 조건식이 있는 list comprehension

`if` 조건을 붙이면 원하는 요소만 필터링해서 리스트를 만들 수 있다.

````python
# 1 부터 100 사이 정수 중 3의 배수만 출력
def make_numbers3() -> list[int]:
    return [lst for lst in range(1, 101) if lst % 3 == 0]


print('3의 배수 리스트 : ', make_numbers3())
````

````python
# 전달된 리스트에서 정수 타입만 필터링해서 반환
lst = [1, 1.1, 2, 3, 4.4, 'aaa', True, False, {}, [], ()]


def fillter_int_values(values: list[object]) -> list[int]:
    # isinstance(value, type) : 결과로 boolean타입을 반환
    return [v for v in values if isinstance(v, int) == True]


print("정수형만 전달 : ", fillter_int_values(lst))
````

## 3. 중첩 반복문 list comprehension

반복문이 중첩된 구조도 comprehension으로 표현할 수 있다. 다만 너무 복잡해지면 일반 반복문이 더 읽기 쉽다.

````python
import random

# 중첩 반복문 버전
print('----- 2중 for문 -----')


def make_positions() -> list[tuple[int, int]]:
    positions: list[tuple[int, int]] = []
    for row in range(1, 5):
        for col in range(5):
            positions.append((row, col))
    return positions


print(make_positions())
# print("랜덤 : " ,make_positions_random())
````

````python
# 중첩 반복문 버전
print('----- 2중 for문 컴프리핸션 -----')


def make_positions_whit_compprehension() -> list[tuple[int, int]]:
    return [(raw, col) for raw in range(3) for col in range(5)]


print(f'이중포문 컴프리핸션 : {make_positions_whit_compprehension()}')
````

## 4. dict comprehension

`dict`도 comprehension으로 만들 수 있다.

```python
{key식: value식 for 요소 in iterable}
```

````python
def make_square_map() -> dict[int, int]:
    return {k: k ** 2 for k in range(1, 6)}


print(make_square_map())
````

````python
print('---- 문자열 리스트를 전달하여 {"문자열" : "문자열길이"} 형태의 list[dict] 반환')
words = ['apple', 'banana', 'orange', 'grape', 'kiwi']


def make_word_length_map_normal(lst: list[str]) -> dict[str, int]:
    print('---- 일반적인 방법 ----')
    result = {}
    for word in lst:
        result[word] = len(word)
    return result


def make_word_length_map(lst: list[str]) -> dict[str, int]:
    print('---- 컴프리핸션 방법 ----')
    return {word: len(word) for word in lst if len(word) >= 5}


print('일반 : ', make_word_length_map_normal(words))
print('컴프리핸션 : ', make_word_length_map(words))
````

## 5. list와 dict 변환

`zip()`과 `dict()`를 함께 사용하면 두 리스트를 하나의 딕셔너리로 묶을 수 있다. 반대로 `dict.items()`와 unpacking을 사용하면 딕셔너리에서 key와 value 묶음을 분리할 수 있다.

````python
def make_age_map() -> dict[str, int]:
    names: list[str] = ['홍길동', '이순신', '신사임당']
    ages: list[int] = [20, 30, 40]
    return dict(zip(names, ages))


print('list를 dict로 변경 (list -> dict) : ', make_age_map())


````

````python
def split_age_map() -> tuple[tuple[str,...], tuple[int, ...]]:
    info: dict[str,int] = {'홍길동': 20, '이순신': 30, '신사임당': 40}
    print(info.items())
    print(*info.items())
    names, ages = zip(*info.items())
    return names, ages

print(f'딕셔너라 -> 튜플 묶음 변환 : ', split_age_map())

print('역으로 다시 해보기')
def split_age_tuple() -> dict[str,int]:
    names: tuple[str, ...] = ('홍길동', '이순신', '신사임당')
    ages: tuple[int, ...] = (20, 30, 40)

    return dict(zip(names, ages))

print('역순 tuple -> dict: ', split_age_tuple() )
````
