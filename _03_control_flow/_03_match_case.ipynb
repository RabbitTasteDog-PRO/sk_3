# match...case

- `match...case`는 값의 패턴에 따라 분기 처리하는 제어문이다.
- 다른 언어의 `switch...case`와 비슷하지만, Python의 `match...case`는 값뿐 아니라 구조 패턴도 검사할 수 있다.
- Python 3.10 이상에서 사용할 수 있다.

## 1. 기본 문법

```python
match 표현식:
    case 값1:
        실행문
    case 값2:
        실행문
    case _:
        기본 실행문
```

`case _`는 앞의 어떤 조건에도 맞지 않을 때 실행되는 기본 분기이다.

````python
import random


# 조건식 : 결과가 bool type인 식
# 표현식 : 결과가 값 1개로 떨어지는 식
def print_drink_price(num: str) -> None:
    match num:
        case '1':
            print("콜라 : 1000원")
        case "2":
            print("사이다 : 900원")
        case "3":
            print("몬스터 : 2000원")
        case _:
            print("잘못 입력하였습니다.")
    print("*** 이용해 주셔서 감사합니다 ***")


print_drink_price(input("1,2,3,4 중 입력 하세요"))
````

## 2. 계산기 분기

연산자처럼 정해진 값 중 하나를 선택하는 경우 `match...case`를 사용하면 분기 구조가 깔끔해진다.

````python
import random


def calculate(a: int, b: int, op: str) -> None:
    match op:
        case "+":
            calc = a + b
        case "-":
            calc = a - b
        case "*":
            calc = a * b
        case "/":
            calc = a / b
        case "//":
            calc = a // b
        case "%":
            calc = a % b
        case "**":
            calc = a ** b
        case _:
            calc = "있을수없는값입니다";
    print(f'{a} {op} {b} = {calc}')


# calculate(1,2,"+")
# calculate(1,2,"-")
# calculate(1,2,"*")
# calculate(1,2,"/")
# calculate(1,2,"//")
# calculate(1,2,"%")
# calculate(1,2,"**")

# for op in ["+", "-", "*", "/", "//", "%", "**", '@']:
#     calculate(1,2,op)

for op in ["+", "-", "*", "/", "//", "%", "**", '@']:
    calculate(random.randint(1, 5), random.randint(1, 5), op)
````

## 3. guard 조건

`case 값 if 조건` 형태를 사용하면 단순 값 비교보다 더 세밀한 조건을 줄 수 있다.

````python
# 양의 정수, 음의 정수, 0판독기
def classfiy_number(n: int) -> None:
    match n:
        case m if n > 0:
            print("양의 정수 입니다")
        case m if n < 0:
            print("음의 정수 입니다")
        case _:
            print("0입니다")


for i in [-1, 0, 1]:
    # classfiy_number(random.randint(-10,10))
    classfiy_number(i)
````

## 4. tuple 패턴 검사

`match...case`는 값 하나뿐 아니라 튜플 구조도 검사할 수 있다. `_`는 해당 자리에 어떤 값이 와도 된다는 뜻이다.

````python
def inspect_tuple(value: tuple[int, int]) -> None:
    match value:
        case (1, 2):
            print("(1,2) 가 전달되었습니다")
        case (1, k) if k < 0:
            print('(1,음수) 가 전달 되었습니다')
        case (1, _):
            print('(1,나머지)가 전달되었습니다')
        case _:
            print('그외의 경우')


# inspect_tuple((1,2))
# inspect_tuple((1,-1))
# inspect_tuple((1,1.1))
# inspect_tuple((2,1))

for v in [(1, 2), (1, -1), (1, 1.1), (2, 1), (1, 2, 3)]:
    inspect_tuple(v)
````

## 5. dict 패턴 검사

딕셔너리도 특정 key와 value 패턴을 기준으로 분기할 수 있다.

````python
applicants = [
    {'type': 'dev', 'career': 15},
    {'type': 'dev', 'career': 5},
    {'type': 'devops', 'career': 0},
    {'type': 'hr'},
]


def inspect_applicant(applicant: dict[str, str | int]) -> None:
    match applicant:
        case {'type': 'dev', 'career': career} if career > 10:
            print("중견 개발자 입니다")
        case {'type': 'dev'}:
            print("일반 개발자 입니다")
        case {'type': 'devops'}:
            print('운영/개발자 지원입니다.')
        case _:
            print("개발 직군이 아닙니다")


for obj in applicants:
    inspect_applicant(obj)
````

## 6. match...case로 학점 계산하기

`if...elif...else`로 작성했던 학점 계산도 `match...case`와 guard 조건으로 표현할 수 있다.

````python
# case문 guard패턴
def get_grade(score: int) -> str:
    # if not (type(score) == int):
    #     raise ValueError("숫자만 입력 가능합니다.")

    # if not (0 >= score >= 100):
    if score < 0 or score > 100:
        raise ValueError("정확한 수치를 입력하세요")
    # 코드 작성
    match score:
        case n if score >= 90:
            grade_ = "A"
        case n if score >= 80:
            grade_ = "B"
        case n if score >= 70:
            grade_ = "C"
        case n if score >= 60:
            grade_ = "D"
        case _:
            grade_ = "F"
    return grade_


try:  # Error가 발생할 수 있는 구문을 작성해서 시도
    raw_score = input("점수 입력 : ")
    score = int(raw_score)
    grade = get_grade(score)
    print(f"{score}점 -> {grade} 등급")

except ValueError as error:  # try에서 ValueError 발생 시 수행
    print(error)
````
