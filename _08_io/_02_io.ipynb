## 파일 쓰기

```python
f = open('helloworld.txt', 'w', encoding='utf-8')
print(f, type(f))

text = '''
뭐야 외 날라갔어
일해라 절해라 하지마
시럽계는 달달하고 입문계는 초보야
'''
f.write(text)

text = open('helloworld2.txt', 'r', encoding='utf-8')

print(text.read())
# 파일 연결에 사용된 메모리 반환
# f.close()
```

```python
# - with절
# - with 옆에 작성된 open() 구문에서 사용한 파일 연결을
#   with절 종료 시 자동으로 close

with open('helloworld2.txt', 'w', encoding='utf-8') as f:
    f.write('with을 이용하면 자동 close()된다')
# print(f.read())

with open('helloworld.txt','r', encoding='utf-8') as f:
    print(f.read())
```

## 파일 읽기

```python
# 한줄씩 읽기
with open('helloworld.txt','r', encoding='utf-8') as f:
    for i,line in enumerate(f):
        print(f'i : {i} / line : {line}', end='') # end='' => print완료 후 마지막 문자 빈칸
```

```python
# 파일 이어쓰기('a', appen)
with open('helloworld.txt', 'a', encoding='utf-8') as f:
    f.write('''ㅋㅋㅋ
ㅎㅎㅎㅎ
ㅌㅌㅌㅌ
ㅊㅊㅊㅊ''')

with open('helloworld.txt','r', encoding='utf-8') as f:
    print(f.read())
```

```python
# 파일복사
# 1. 파일일기
with open('helloworld.txt', 'r', encoding='utf-8') as f:
    text = f.read()
# 2. 파일쓰기
with open('helloworld_copy.txt', 'w', encoding='utf-8') as f:
    f.write(text)
    print("파일 복사 완료")
```

```python
# 이미지 파일 복사(아무 파일이나 가능)
# 문자는 : r,w,a는 문자열만 가능
# 문자가 아닌 파일들은: rb,wb 바이트단위
with open('image_test.png','rb') as img:
    imb_bites = img.read()
    # print(img.read())
with open('image_test_copy.png','wb') as img:
    img.write(imb_bites)
    # print('이미지 복사 완료')
```
