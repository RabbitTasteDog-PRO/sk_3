### **문제 7**

아래 조건에 맞게 리스트의 요소를 수정하세요.

1. `colors` 리스트에 `'red', 'blue', 'green'`을 저장하세요.
2. `'blue'`를 `'yellow'`로 변경하고 리스트를 출력하세요.

```python
colors: list[str] = ['red', 'blue', 'green']
print(colors)
# for color in colors:
for i, color in enumerate(colors):
    if color == 'blue':
        colors[i] = 'yellow'
        break
print(colors)

print('---- indexing ----')
colors2: list[str] = ['red', 'blue', 'green']
print(f'원본 : {colors2}')
idx = colors2.index('blue')
colors2[idx] = 'yellow'
print(f'변경 : {colors2}')
```

### **문제 8**

아래 조건에 맞는 리스트 내포를 작성하세요.

1. 숫자 `1`부터 `10`까지 중에서 **짝수만**을 추출해 새로운 리스트에 저장하세요.
2. 리스트를 출력하세요.

```python
lst: list[int] = []
for i in range(1, 11):
    lst.append(i)
print(lst)

lst2: list[int] = []
for i in lst:
    if i % 2 == 0:
        lst2.append(i)

print(lst2)
```

### **문제 9**

아래 조건에 맞게 리스트와 문자열을 활용하세요.

1. 리스트 `words`에 `'Life', 'is', 'too', 'short'`를 저장하세요.
2. 리스트의 요소를 공백으로 이어붙여 문자열 `'Life is too short'`를 출력하세요.

```python
words: list[str] = ['Life', 'is', 'too', 'short']

result = ""
for word in words:
    result += (word + " ")
print(result.strip())

print(" ".join(words))
```
