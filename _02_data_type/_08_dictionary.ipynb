## 1. dict 생성과 key 규칙

`dict`의 핵심은 `key`이다.

- 하나의 `dict` 안에서 `key`는 중복될 수 없다.
- 같은 `key`를 다시 넣으면 마지막 값으로 덮어써진다.
- `key`는 값을 빠르게 찾기 위한 기준이므로 중간에 값이 바뀌면 안 된다.
- 그래서 `int`, `float`, `str`, `tuple` 같은 불변 자료형은 `key`로 사용할 수 있지만, `list`, `dict`, `set` 같은 가변 자료형은 `key`로 사용할 수 없다.
- Python 3.7 이상에서 `dict`는 삽입 순서를 유지한다.dict(Dictionary)

```python
dic3 = dict(name ="홍길동", age =20)
print(dic3)
print(f'key 들만 나오게 : {dic3.keys()} , type : {type(dic3.keys())}')
print(f'value 들만 나오게 : {dic3.values()}, type : {type(dic3.values())}')
print(f'items 들만 나오게 : {dic3.items()}, type : {type(dic3.items())}')
```
