```python
# except 구분을 이용한 예외 객체 사용 방법
import random


def try_exception(e: Exception, msg: str) -> None:
    if msg == '':
        print(f'[Error] {msg}')
    else:
        print(e)
        print("예외 타입 이름 ", type(e).__name__)
        print("예외 메시지 ", str(e))
        print("예외 상세값 ", str(e.args))
        print(f'[Error] {type(e).__name__} : {str(e)}')


try:
    num = random.randint(0, 1)
    print(num)
    # 0 == false, 1 == true
    if num:  # 일때
        print('abc' + 123)  # typeError
    else:  # 0일때
        print(int('abc'))  # ValueError
except TypeError as e:
    try_exception(e)
except ValueError as e:
    try_exception(e)



```

### pass
- if, for, def, class, except 등의 블록은 내부가 비어있으면 오류가 발생함
- 이 때, 구조만 유지하는 용도 또는 예외를 무시할 때 사용

```python
if True:
    try:
        print(10 / 0)
    except Exception as e:
        print(f"pass error : {type(e).__name__}")
        pass


def func():
    pass
```

### finally
- try 구문에서 예외 발생 여부와 관계없이 마지막에 무조건 실행하게 하는 구문(블록)

### else
- except 구문을 모두 지나친 경우 수행되는 구문


```python
try:
    num = int(input("숫자 입력"))
    result = 100 / num
except ValueError as e:
    print('정수형 숫자만 입력하세요')
except ZeroDivisionError as e:
    print('0으로 나눌 수 없습니다')
else:
    print("계산결과 : ", result)
finally:
    print("예외 발생 여부와 관계없이 무조건 실행")
```

```python
# finally 사용 예시
try:
    f = open('helloworld.txt', 'r', encoding='utf-8')
    print(10 / 0)
except Exception as e:
    pass
finally:
    # 메모리 반환 필수
    f.close()

# 위드절 사용안했다면 f.close() 필수

```
