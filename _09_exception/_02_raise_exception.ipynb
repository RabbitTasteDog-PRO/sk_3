```python
# 사용자 정의 예외 클래스
# - 만드는 방법 : 기존 Exception, Error 클래스 중 아미거나 상속 받기
class UnderException(Exception):
    pass


class AdultGame2():
    def __init__(self, age: int):
        if age < 19:
            # BaseException : 모든 예외 에러 클래스의 최상위 부모 클래스
            # raise BaseException(f'미성년자는 이용할 수 없습니다: {age}')
            # 파이썬에 제공하는 알맞은 예외가 없다면, 사용자가 생성 가능하다
            raise UnderException(f'미성년자는 이용할 수 없습니다: {age}')

    def start(self):
        print('성인용 게임(고스톱)을 시작합니다.')


try:
    age = int(input('나이를 입력하세요 :'))
    game2 = AdultGame2(age)
    game2.start()
except UnderException as e:
    print("error : ", type(e).__name__, "message : ", str(e))

print('종료')
```
