## map
iterable객체의 각요소에 대해 지정한 함수를 적용하고, 그 결과를 map객체로 반환
```
# iterable : 반복 가능한 요소들
map(function, iterable) -> map
```




````python
from traceback import print_tb

# map() : 기존 list의 요소를 하나씩 꺼내 함수를 호출하고 함수결과들을 모아 새로운 리스트를 생성
nums: list[int] = [1, 2, 3, 4, 5]
result = map(lambda x: x ** 2, nums)
print(f"reuslt : {result}, type : {type(result)}")
print('list(result) : ', list(result))
````

````python
# 문자열 리스트 -> 문자열 길이 리스트
````

````python
products: list[str] = ['아이스아메리카노', '수박주스', '말차라떼', '식혜', '맥콜']
# products_len = list[int] = map(콜백함수, 순회가능한 객체(list, dict, set, tuple))
products_len: list[int] = map(lambda x: len(x), products)
print('리스트 변경 products_len', list(products_len))
````

````python
#여러 순회가능 객체(list) 통합
````

````python
mid_term = [90, 80, 77, 99, 88, 33]
final_term = [100, 99, 11, 22, 44]

avg_scores = list(map(lambda x, y: (x + y) / 2, mid_term, final_term))
print(f'평균 : {avg_scores}')
````

## filter

주어진 iterable객체에 대해서 참/거짓 평가후 참인 요소들만 결과값으로 반환

```
filter(function, iterable) -> filter
```

````python
nums: list[int] = list(range(1, 11))
evens = filter(lambda x: x % 2 == 0, nums)
odds = filter(lambda x: x % 2 != 0, nums)
threes = filter(lambda x: x % 3 == 0, nums)
print(f'짝수모음 : {evens}, type : {type(evens)}')
print(f'짝수모음  : {list(evens)}, type : {type(evens)}')

print('홀수모음 : ', list(odds))
print('3의 배수 모음 : ', list(threes))
````

````python
# 이름 목록에서 성이 "홍" 씨인 사람만 찾아서 list 반환
names: list[str] = ['홍길동', '고길동', '홍진경', '최홍만', '홍명보']

name1 = list(filter(lambda x: x[0] == "홍", names))
name2 = list(filter(lambda x: x.startswith('홍'), names))
print(name1)
print(name2)
````

````python
#dictionary 검색 (필터링)
# 학생 목록에서 점수가 90점 이상인 학생만 필터링
students = [
    {'name': '이순신', 'score': 80},
    {'name': '윤봉길', 'score': 75},
    {'name': '논개', 'score': 99},
    {'name': '강감찬', 'score': 91},
]

student1 = list(filter(lambda x: x['score'] >= 90, students))
print('90점 이상인 학생 : ', student1)

# list comprehension
student2 = [s for s in students if s['score'] >= 90]
print('컴프리핸션 리스트 결과 : ', student2)
````

## reduce
iterable객체의 요소에 대한 연산결과를 단일값을 반환.
합계, 최대/최소값 등의 연산에 적합

```
reduce(function, iterable, initial_value) -> Any
```
함수는 3개의 인자를 받는다.
1. 누적 연산값
2. 현재 요소
3. 초기값 (선택)


````python
# 외부 펑션툴스의 리듀스 기능을 가져옴
# 리듀스는 기본 내장이 아님
from functools import reduce

# 정수형 list 내 요소들의 합계
nums: list[int] = [11, 22, 33, 44, 55, 66, 77, 88, 99]

#총합 저장
# totla = reduce(콜백함수(누적용, 현재값), 이터러블값(반복값:리스트나 딕셔너리), 누적변수초기값)
total = reduce(lambda sum, cur: sum + cur, nums, 0)
# 콜백 함수의 return 값이 자동으로 누적용 변수에 대입 된다
print("리듀스 함수 : ", total)
print("일반 sum : ", sum(nums))

````

````python
# reduce를 이용한 최대,최소값 찾기
nums: list[int] = [1, 2, 3, 34, 4, 5, 5, -6, 6, 4, 2, 34, 3, -10000000]
maxNum: int = reduce(lambda max, crr: max if max > crr else crr, nums, 0)
print('maxNum', maxNum)
print("max함수 : ", max(nums))
nimNum: int = reduce(lambda min, crr: min if min < crr else crr, nums, 0)
print('nimNum', nimNum)
print("min함수 : ", min(nums))


# 절대값이 가장 큰 수
def abs_func(max, crr) -> int:
    print(f'max : {abs(max)}, crr : {abs(crr)}')
    return max if abs(max) > abs(crr) else crr


abs_max = reduce(abs_func, nums, 0)
print('abs_max', abs_max)
````

````python
# 최고 점수를 지닌 학생 출력
students = [
    {'name': '이순신', 'score': 80},
    {'name': '윤봉길', 'score': 75},
    {'name': '논개', 'score': 99},
    {'name': '강감찬', 'score': 91},
]

# dic['id']
student = reduce(lambda max, crr: max if max['score'] > crr['score'] else crr, students, {'name': '', 'score': 0})
# dic.get
student2 = reduce(lambda max, crr: max if max.get('score') > crr.get('score') else crr, students, {'name': '', 'score': 0})
# 초기값 생략 가능
student3 = reduce(lambda max, crr: max if max.get('score') > crr.get('score') else crr, students)
print('student',student)
print('student2',student2)
print('student3',student3)

````
