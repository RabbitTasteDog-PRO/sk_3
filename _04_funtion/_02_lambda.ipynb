# lambda 람다

- `lambda`는 함수를 한 줄로 간결하게 선언하는 문법이다.
- 이름이 없는 작은 함수를 만들 때 사용한다.
- 기본 구조는 `lambda 매개변수: 반환값`이다.
- 복잡한 로직은 일반 `def` 함수로 작성하는 편이 읽기 쉽다.
```
변수명 = lambda (매개변수): (함수 실행 내용)
# 함수 실행 내용에 작성된 코드(표현식)이 끝나면 자동으로 반환된다
```

## 1. lambda 기본 문법

아래 두 코드는 같은 의미이다.

```python
def square(x):
    return x ** 2

square = lambda x: x ** 2
```

````python
print('--- 일반 함수 ----')


def square(x: int) -> int:
    return x ** 2


print(square(2))

print('---- 람다 함수 ----')
square2 = lambda x: x ** 2
print(square2(5))
````

## 2. list comprehension 안에서 사용하기

`lambda`는 짧은 계산식을 값처럼 전달하거나 즉시 실행할 때 사용할 수 있다.

````python
def make_squears() -> list[int]:
    return [n for n in range(1, 11)]


print("일반 컴프리헨스 : ", make_squears())


def make_lambda() -> list[int]:
    return [(lambda x: x ** 5)(n) for n in range(1, 11)]


print("람다식 컴프리헨스 : ", make_lambda())

````

## 3. 여러 매개변수 사용

`lambda`도 일반 함수처럼 하나 이상의 매개변수를 받을 수 있다.

````python
multy = lambda a, b: a * b
print('multy : ', multy(2, 3))

add = lambda a, b: a + b
print('add : ', add(2, 3))
````

## 4. 기본값, `*args` 사용

`lambda`에서도 기본값이 있는 매개변수와 `*args`를 사용할 수 있다. 다만 코드가 길어지면 `def` 함수로 분리하는 것이 좋다.

````python
print('---- lambda 매개변수 기본값 ----')
multyply = lambda a, b=2: a * b
print(multyply(10))

print('---- lambda 매개변수 *args ----')
adder = lambda *args: sum(args)
print(adder(1, 2, 3, 4, 5))
````

## 5. lambda와 고차 함수

고차 함수는 함수를 인자로 받거나, 함수를 반환하는 함수이다.

`lambda`는 이름 없는 함수를 값처럼 만들어서 다른 함수에 전달할 때 자주 사용한다.
아래 예제에서 `apply_to_numbers()`는 `func`라는 함수를 인자로 받는 고차 함수이다.


````python
numbers: list[int] = [1, 2, 3, 4, 5]


def apply_to_numbers(nums, func) -> list[int]:
    result = []
    for i in nums:
        result.append(func(i))
    return result


def function2(x):
    return x * 2


# 일반 람다
numbers1 = apply_to_numbers(numbers, lambda x: x * 2)
# 람다 풀어서
numbers2 = apply_to_numbers(numbers, function2)
# 람다 제곱
numbers3 = apply_to_numbers(numbers, lambda x: x ** 2)
# 컴프리헨스 & 람다
numbers4 = apply_to_numbers([n for n in range(1, 11, 2)], lambda x: x - 1)

print(numbers1)
print(numbers2)
print(numbers3)
print(numbers4)
````

````python
# 함수를 반환하는 고차 함수
def make_multyplier(n: int):
    return lambda number: number * n


triple = make_multyplier(3)
print(triple(5))

print(f'lambda : {(lambda x: x * 2)(10)}')
````

## 6. lambda와 콜백 함수

콜백 함수는 내가 바로 실행하는 함수가 아니라, 다른 함수에 전달해 두었다가 필요한 시점에 호출되는 함수이다.

`lambda`는 콜백 로직이 짧을 때 사용하기 좋다. 다만 콜백 안의 코드가 길어지면 이름 있는 `def` 함수로 작성하는 편이 읽기 쉽다.


````python
def calc(a, b, operation):
    return operation(a, b)


add_result = calc(999, 888, lambda x, y: x + y)
print('add_result : ',add_result)

minus_result = calc(999,888, lambda z,w: z-w)
print('minus_result : ',minus_result)
````
