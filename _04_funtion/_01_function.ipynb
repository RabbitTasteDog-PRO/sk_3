# Function 함수

특정 기능을 수행하기 위한 코드의 집합을 의미한다.
함수를 정의해 놓으면 해당 기능이 필요한 위치에서 함수를 호출하여 간편하게 사용할 수 있다.

- 함수는 반복되거나 절차적인 실행 코드를 하나의 이름으로 묶은 것이다.
- 함수를 선언한 뒤 호출해서 사용한다.
- 호출할 때 값을 전달할 수 있고, 실행 결과를 반환받을 수 있다.
- Python에서 함수는 하나의 값처럼 다룰 수 있는 1급 객체이다.

## 1. 함수 선언과 호출

`def` 키워드로 함수를 선언하고, 함수 이름 뒤에 괄호를 붙여 호출한다.

```python
from tkinter.font import names


# 값 전달 : O, 결과 반환 : O
# 함수 선언
def add(a, b):
    # 함수 정의
    return a + b


# 함수 호출
print(f'add 함수 호출 : ', add(10, 20))
print(type(add))


# 값 전달 : O, 결과 반환 : X
def minus(a, b):
    print(f' {a} - {b} : {a - b}')


minus(10, 20)


# 값 전달 : X, 결과 반환 : O
def multyply():
    return 111 * 111


print(multyply())


# 값 전달 : X, 결과 반환 : X
def greeting():
    print("hello hell")


greeting()
```

## 2. 매개변수와 인수

- **매개변수(parameter)**: 함수 선언부에서 값을 받기 위해 만든 변수
- **인수(argument)**: 함수 호출부에서 실제로 전달하는 값

```python
# 매개변수 == 파라미터
def foo(k, m):
    print("k : ", k)
    print("m : ", m)


# 인자값 ( arguments)
foo(100, 200)
```

## 3. return과 함수 객체

함수는 실행을 마치면 호출한 위치로 돌아온다. 이때 `return`으로 값을 하나 돌려줄 수 있다.

또한 Python에서 함수는 변수에 담거나 다른 함수의 인수로 전달할 수 있다. => 1급 개체라 한다

### 1급 객체?
- 값처럼 다룰수 있는 객체
    1. 변수에 대입 가능
    2. 함수 인자로 전달 가능
    3. 함수 반환 값으로 사용 가능
    4. 자료구조에 저장 가능

```python
def bar():
    # print('---- 일반적인 데이터 타입 return ----')
    print("bar")
    return 0


# print('---- bar() 함수를 func변수에 담는다 ----')
func = bar
# func2 = bar()

print(f'bar() 주소 : {id(bar)}')
print(f'func() 주소 : {id(func)}')
print(f'bar is func : {bar is func}')


def runner(f):
    print(f())


runner(func)
```

## 4. 위치 인수와 키워드 인수

함수를 호출할 때 인수를 전달하는 방식은 크게 두 가지이다.

- 위치 인수: 작성한 순서대로 매개변수에 전달
- 키워드 인수: `key=value` 형태로 같은 이름의 매개변수에 전달

두 방식을 섞을 때는 위치 인수를 먼저 쓰고, 그 뒤에 키워드 인수를 써야 한다.

```python
def test(a, b):
    print('a:', a)
    print('b:', b)


print('---- 위치인수 ----')
test(10, 20)

print('---- 키워드 인수 ----')
test(a=30, b=40)
test(b=30, a=40)

print('---- 위치 인수 + 키워드 인수 ')
test(100, b=200)
# test(a=100, 200)
# 키워드 인수 뒤에는 위치 인수를 작성 할 수 없다
```

## 5. 매개변수 기본값

기본값이 있는 매개변수는 호출할 때 생략할 수 있다.

단, 기본값이 있는 매개변수는 기본값이 없는 매개변수보다 오른쪽에 작성해야 한다.

```python
# 인수 누락시 기본값 적용
def test2(name, age, address="default"):
    print(f'name : {name}, age : {age}, address : {address}')


test2('홍길동', 20)

#잘못된 경우
# def wrong(name, age=30, address):
#     print(f'name : {name}, age : {age}, address : {address}')
```

## 6. packing: `*args`, `**kwargs`

- `*args`: 여러 위치 인수를 하나의 튜플로 모은다.
- `**kwargs`: 여러 키워드 인수를 하나의 딕셔너리로 모은다.
- 매개변수 작성 순서는 일반 매개변수, `*args`, `**kwargs` 순서를 기본으로 생각하면 좋다.

```python
# 전달받은 인자를 튜플로 묶음
def test3(*args):
    print(f'arge : {args}')
    return sum(args)


print(f'튜플로 묶임 : {test3(1, 2, 3)}')
print(f'튜플로 묶임 : {test3(1, 2, 3, 4, 5, 6, 6, 8)}')


# 기본인수 + *args

def test4(emoji, *arge):
    print(f'emoji : {emoji}, arge : {arge}')


test4("👍", 1, 2, 3, 3)
test4(("👍" * 3), 1, 2, 3, 5)
test4("👍")
```

```python
def test5(**kwargs):
    print(f'kwargs : {kwargs}')
    print(f'type : {type(kwargs)}')


# dict 형식 확인
test5()

test5(menu="돈까스", price=10000, options=['치즈', '콜라'])


# 기본인수 + **kwargs
def test6(name, age, **kwargs):
    print(f'name : {name}, age : {age}, kwargs : {kwargs}')


test6('홍길동', 20, a=10, b=20, hobby=["게임", "공부"])


# 일반인수 + *args + **kwargs
def test7(a, b, *args, **kwargs):
    print()
    print(f'a : {a}, b : {b}')
    print(f'args : {args}')
    print(f'kwargs : {kwargs}')
    # print(f'a : {a}, b : {b}, args : {args}, kwargs : {kwargs}')


test7(10, "a", 1, 2, 3, 4, "aa", 1.1, [], {}, menu="aaa", cc=10, dd="", ee=["aaa", (1, 2), [1, 2, 3]])
```

## 7. unpacking: `*list`, `**dict`

- `*list`는 리스트의 요소를 꺼내어 위치 인수처럼 나열한다.
- `**dict`는 딕셔너리의 key-value를 꺼내어 키워드 인수처럼 나열한다.

```python
names = ['홍길동', '이순신', '카리나']


def test8(*args):
    print(args)


print('---- list unpacking ----')
print('---- 아래는 리스트로 묶은상태 ----')
test8(names)
print('---- 아래는 리스트를 unpacking ----')
test8(*names)
print('---- 리스트를 풀어서 각 요소 인자로 전달 ----')
```

```python
info = {
    'date': '2026/06/04',
    'time': '10:00',
}


def test9(date, time):
    print(f'date : {date}')
    print(f'time : {time}')


print('---- dict 일반 키 값 ----')
test9(info['date'], info['time'])
print('---- dict 키 인자  ----')
test9(time=info['time'], date=info['date'])
print('---- dict 언패킹 ----')
test9(**info)

```

## 8. 위치 전용과 키워드 전용 매개변수

Python 함수는 특정 매개변수를 위치 전용 또는 키워드 전용으로 제한할 수 있다.

- `/` 앞의 매개변수는 위치 전용이다.
- `*` 뒤의 매개변수는 키워드 전용이다.

```python
# / 앞에는 매개변수만 허용한다
def test10(name, /, gretting="안녕"):
    print(name, gretting)


test10('홍길동')
# 위치인수 전용이라 키워드는 동작하지 않고 에러
# test10(name='홍길동', gretting='hello')
test10('홍길동', gretting='hello')


# * 뒤에는 키워드 인수만 허용한다
def test11(a, *, b, c):
    print(a, b, c)


test11(10, c=20, b=30)
```

## 9. 함수 type hint

함수의 매개변수와 반환값에도 type hint를 작성할 수 있다. type hint는 실행을 강제하기보다 IDE와 정적 분석 도구가 코드를 이해하도록 돕는 역할을 한다.

```python
# 매개변수 타입 힌트  == a: str,b: int,c: bool
# 반환형 타입힌트 == -> dict[str,object]
def test12(a: str, b: int, c: bool) -> dict[str, object]:
    return {'a': a, 'b': b, 'c': c}

print(test12('홍길동',20, True))

```
